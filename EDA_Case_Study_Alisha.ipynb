```python
# Check data types of columns having missing values less than 16%
print('NAME_TYPE_SUITE', application_data1['NAME_TYPE_SUITE'].dtypes, 'Method = Mode', application_data1['NAME_TYPE_SUITE'].mode()[0])
print('AMT_REQ_CREDIT_BUREAU_HOUR', application_data1['AMT_REQ_CREDIT_BUREAU_HOUR'].dtypes, 'Method = Mean', application_data1['AMT_REQ_CREDIT_BUREAU_HOUR'].mean() )
print('AMT_REQ_CREDIT_BUREAU_DAY', application_data1['AMT_REQ_CREDIT_BUREAU_DAY'].dtypes, 'Method = Mean', application_data1['AMT_REQ_CREDIT_BUREAU_DAY'].mean() )
print('AMT_REQ_CREDIT_BUREAU_WEEK', application_data1['AMT_REQ_CREDIT_BUREAU_WEEK'].dtypes, 'Method = Mean', application_data1['AMT_REQ_CREDIT_BUREAU_WEEK'].mean() )
print('AMT_REQ_CREDIT_BUREAU_MON', application_data1['AMT_REQ_CREDIT_BUREAU_MON'].dtypes, 'Method = Mean', application_data1['AMT_REQ_CREDIT_BUREAU_MON'].mean() )
print('AMT_REQ_CREDIT_BUREAU_QRT', application_data1['AMT_REQ_CREDIT_BUREAU_QRT'].dtypes, 'Method = Mean', application_data1['AMT_REQ_CREDIT_BUREAU_QRT'].mean() )
print('AMT_REQ_CREDIT_BUREAU_YEAR', application_data1['AMT_REQ_CREDIT_BUREAU_YEAR'].dtypes, 'Method = Mean', application_data1['AMT_REQ_CREDIT_BUREAU_YEAR'].mean() )

# We can replace the missing values with below mentioned values. Since NAME_TYPE_SUITE is a categorical variable, we have used
# mode to replace values. For others, we have used mean as the values are continuous in nature 
```

```python
#application_data1.NAME_TYPE_SUITE.fillna(application_data1.NAME_TYPE_SUITE. mode(), inplace=True)
application_data1.AMT_REQ_CREDIT_BUREAU_HOUR.fillna(application_data1.AMT_REQ_CREDIT_BUREAU_HOUR.mean(), inplace=True)
application_data1.AMT_REQ_CREDIT_BUREAU_DAY.fillna(application_data1.AMT_REQ_CREDIT_BUREAU_DAY.mean(), inplace=True)
application_data1.AMT_REQ_CREDIT_BUREAU_WEEK.fillna(application_data1.AMT_REQ_CREDIT_BUREAU_WEEK.mean(), inplace=True)
application_data1.AMT_REQ_CREDIT_BUREAU_MON.fillna(application_data1.AMT_REQ_CREDIT_BUREAU_MON.mean(), inplace=True)
application_data1.AMT_REQ_CREDIT_BUREAU_QRT.fillna(application_data1.AMT_REQ_CREDIT_BUREAU_QRT.mean(), inplace=True)
application_data1.AMT_REQ_CREDIT_BUREAU_YEAR.fillna(application_data1.AMT_REQ_CREDIT_BUREAU_YEAR.mean(), inplace=True)
# Finding percentage of Nan
application_data1_percent = round(100*(application_data1.isnull().sum()/len(application_data1.index)),2)
application_data1_percent
```

```python
# #plotting graph for few numerical and categorical columns for univariate analysis for merged datset
def plotGraph(x, y):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    sns.countplot(x = x, data  = y)

arr = ['AMT_REQ_CREDIT_BUREAU_HOUR','AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON','AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR','CNT_CHILDREN','CODE_GENDER','FLAG_OWN_CAR','FLAG_OWN_REALTY']

for i in arr:
    plotGraph(i,application_prev_data)
```
